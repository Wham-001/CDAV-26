write a program that takes a value from the user. prog should print
#divisible by two if the value is divisible by 2
#divisible by three if the value is divisible by 3
#divisible by both 2 and 3 if value is divisible by both 2 and 3
#max 2 statements
#otherwise it should print not divisible by both 2 and 3

```python
value=input("enter a value ")
value=int(value)
is_divisible_by_2=value/2
is_divisible_by_3=value/3
is_divisible_by_2_and_is_divisible_by_3=value/2 and value/3
if is_divisible_by_2==("divisible by 2"):
  print("divisible by 2")
elif is_divisible_by_3==("divisible by 3"):
  print("divisible by 3")
elif is_divisible_by_2 and is_divisible_by_3==("divisible by 2 and 3"):
  print("divisible by 2 and 3")
else:
  print("not divisible by 2 and 3")
```
