```python
letter=input("enter letter of the alphabet ")
is_vowel=letter in "aeiou"
is_vowel_and_consonant=letter == "y"
if is_vowel:
  print("letter entered is vowel")
elif is_vowel_and_consonant:
  print("letter entered is vowel and consonant")
else:
  print("letter entered is consonant")
```

```python
sound=int(input("sound level in decibels "))
is_jackhammer=sound==130
is_gas_lawnmower=sound==106
is_alarm_clock=sound==70
is_quiet_room=sound==40
if is_jackhammer:
  print("sound level is jackhammer")
elif is_gas_lawnmower:
  print("sound level is gas lawnmower")
elif is_alarm_clock:
  print("sound level is alarm clock")
elif is_quiet_room:
  print("sound level is quiet room")
elif sound<40:
  print("sound level is less than quiet room")
elif sound>130:
  print("sound level is greater than jackhammer")
elif sound>40 and sound<70:
  print("sound level is between quiet room and alarm clock")
elif sound>70 and sound<106:
  print("sound level is between alarm clock and gas lawnmower")
elif sound>106 and sound<130:
  print("sound level is between gas lawnmower and jackhammer")
```
