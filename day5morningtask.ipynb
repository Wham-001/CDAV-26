GCD (question 75)


```python
n=int(input("enter first integer: "))
m=int(input("enter secong integer: "))
d=min(n,m)
while m%d!=0 or n%d!=0:
  d=d-1
print(f"The greatest common divisor of {n} and {m} is {d}.")

```

Palindrome(question 72)

```python
text = input("Enter a string: ")
left = 0
right = len(text) - 1
is_palindrome = True
while left < right:
    if text[left] != text[right]:
        is_palindrome = False
        break  #manual way of ending the loop instead of a control
    elif text[left] == text[right]:
        left += 1
        right -= 1
if is_palindrome:
    print((text),"is a palindrome")
elif not is_palindrome:
    print((text),"is NOT a palindrome")
```

Caeasar cypher(question 70)

```python

message = input("Enter your message: ")
shift = int(input("Enter the shift amount (use a negative number to decode): "))

result = ""

for letter in message:

    if letter.isupper():
        position = ord(letter) - ord('A')
        new_position = (position + shift) % 26
        new_letter = chr(new_position + ord('A'))
        result = result + new_letter

    elif letter.islower():
        position = ord(letter) - ord('a')
        new_position = (position + shift) % 26
        new_letter = chr(new_position + ord('a'))
        result = result + new_letter

    else:
        result = result + letter

print("Your message was:", message)
print("Your shifted message is:", result)
```
