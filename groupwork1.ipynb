Student Record Management System

This system allows you to input student details and then provides options to:

View all student records, including their pass/fail status.

View only students who failed (score below 50).

Exit the program.

All student data is stored efficiently using dictionaries.

```python
#creates a central database 'container' for the program
student_record={}

#creates loop for entry of multiple student details,marks etc
while True:
  print("1.Add details")
  print("2.Record marks")
  print("3.View failing scores")
  print("4.View complete details")
  print("5.Exit program")
  choice=input("enter choice(1-5) ")

  if choice=="1":
    student=input("enter student name ")
    age=int(input("enter age "))
    scores=[] #creates an empty score list
    # student_record initializes/calls the dictionary while [student] acts as
    # the key and [age,scores] are the values
    student_record[student]=[age,scores]
    print(student_record)

  elif choice=="2":
    student=input("enter student name ")
    # checks if the string stored in student exists as a key inside
    # student_record
    if student in student_record:
      #if statement is to check whether the student actually exists
      score=int(input("enter student score "))
      student_record[student][1].append(score)
      print(student_record)
  elif choice=="3":
    print("Failing scores(<50)")
    for student in student_record:
      #get age and sccore from dictionary by using indexes,one at a time
      age=student_record[student][0]
      scores=student_record[student][1]
      for score in scores:
        if score <50:
          print(f"Student:{student} Age{age} Score:{score} Fail")
        else:
          print(f"Student:{student} Age:{age} Score:{score} Pass")
  elif choice=="4":
    print("Complete details")
    for student in student_record:
      age=student_record[student][0]
      scores=student_record[student][1]
      if scores and scores[-1]<50:
        status="Fail"
      else:
         status ="Pass"
      print(f"Student:{student} Age:{age} Score:{score} Status:{status}")
  elif choice=="5":
        print("End of program")
        break
```
